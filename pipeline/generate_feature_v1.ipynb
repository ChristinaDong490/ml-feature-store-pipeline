```python
import pandas as pd
```

```python
df_cleaned = pd.read_csv("/Users/christinadong/Documents/MLops/ml-feature-store-pipeline/data/athletes_clean.csv")
```

```python
df_cleaned.head(5)
```

```python
# Feature Version 1
df_v1 = df_cleaned[['age', 'height', 'weight', 'gender', 'deadlift', 'candj', 'snatch', 'backsq']].copy()

# Convert gender to binary
df_v1['gender'] = df_v1['gender'].map({'Male': 1, 'Female': 0})

# Add required Feast timestamp column
import datetime
df_v1['event_timestamp'] = datetime.datetime.utcnow()

# Add dummy ID for Feast joining (required even if not in original)
df_v1 = df_v1.reset_index(drop=True)
df_v1['athlete_id'] = df_v1.index.astype(str)

# Save
df_v1.to_csv('data/feature_v1.csv', index=False)
```
