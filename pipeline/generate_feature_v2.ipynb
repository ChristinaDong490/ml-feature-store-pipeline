```python
df_v2 = df_v1.copy()

# Add new feature BMI
df_v2['BMI'] = df_v2['weight'] / (df_v2['height'] * 0.0254)**2  # assuming inches -> meters

# Add age group feature
df_v2['age_group'] = pd.cut(df_v2['age'], bins=[0, 20, 30, 40, 100], labels=[1, 2, 3, 4])

df_v2.to_csv('/Users/christinadong/Documents/MLops/ml-feature-store-pipeline/data/feature_v2.csv', index=False)
```
